# Who are REG?

```python
import numpy as np

from github_analyser.org_user_info import get_org_members, get_org_teams
from github_analyser.team_user_info import get_team_members
from github_analyser.repo_user_info import get_repo_collaborators
```

## Who are the Alan Turing Institute?

```python
ati_members = get_org_members("alan-turing-institute")
```

```python
print(f"There are {len(ati_members)} users in the Alan Turing Institute GitHub org.")
```

```python
role_counts = ati_members.aggregate("role").value_counts()
```

```python
print(f"Of these users, {role_counts["ADMIN"]} are admins and the other {role_counts['MEMBER']} are members.")
```

```python
ati_teams = get_org_teams("alan-turing-institute")
```

```python
print(f"There are {len(ati_teams)} teams in the Alan Turing Institute GitHub org.")
```

```python
print("Examples of teams in the Alan Turing Institute GitHub org are:")

for i in np.random.choice(ati_teams.index, 10):
    print(f"  - {ati_teams.loc[i, 'name']}")
```

```python
team_members = {}

for i in ati_teams.index:
    team_name = ati_teams.loc[i, "name"]
    team_slug = ati_teams.loc[i, "slug"]
    team_members[team_name] = get_team_members("alan-turing-institute", team_slug)
```

## Who are REG?

```python
print(f"Theoretically, all REG members should be members of the 'Hut23' team.")
```

```python
print(f"There are {len(team_members["Hut23"])} members in the 'Hut23' team.")
```

```python
hut23_collaborators = get_repo_collaborators("alan-turing-institute", "hut23")
```

```python
print(f"There are {len(hut23_collaborators)} collaborators on the Hut23 repository.")
```

```python
hut23_team_collaborators = hut23_collaborators[
    hut23_collaborators["login"].isin(team_members["Hut23"]["login"])
]

non_hut23_team_collaborators = hut23_collaborators[
    ~hut23_collaborators["login"].isin(team_members["Hut23"]["login"])
]
```

```python
assert len(hut23_team_collaborators) == len(team_members["Hut23"])

print(f"All {len(team_members["Hut23"])} 'Hut23' team members are collaborators on the Hut23 repository.")
```

```python
print(
    f"There are {len(non_hut23_team_collaborators)} collaborators on the Hut23 respository who are not in the 'Hut23' team."
)
```

```python
non_turing_collaborators = hut23_collaborators[
    ~hut23_collaborators["login"].isin(ati_members["login"])
]
```

```python
print(
    f"But there is only {len(non_turing_collaborators)} collaborator on the Hut23 repository who is not in the Alan Turing Institute GitHub org."
)
```
